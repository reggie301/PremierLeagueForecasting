```python
from bs4 import BeautifulSoup
import time
import os
from selenium import webdriver
import pandas as pd
from datetime import date
import requests
```

```python
### SPI SCRAPER FOR THIS MATCHWEEK


current_season = int(input('What is the current season?'))


url = r'https://projects.fivethirtyeight.com/soccer-predictions/premier-league/'

driver = webdriver.Chrome()

r = driver.get(url) # proxies=proxies
time.sleep(2)

date_dict = {'Sept':'Sep','March':'Mar','April':'Apr','June':'Jun','July':'Jul'}
  
if not os.getcwd().endswith('Football Forecasting Version 2'):
    os.chdir(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))

current_folder = os.getcwd()

relative_folder = fr'{current_season}_{current_season+1}_SPIdata'   

full_path = os.path.join(current_folder, relative_folder)

isExist = os.path.exists(full_path)


if not isExist:
    os.makedirs(full_path)

match_date = date.today().strftime('%y%b%d')

dfs = pd.read_html(driver.page_source)
a = dfs[0].replace(' ',':')
a.columns = a.columns.droplevel()
a = a[['team','spi','off.','def.']]
a['team'] = a['team'].str.replace('\d+', '',regex = True)
a['team'] = a['team'].str.replace('pts', '')
a['team'] = a['team'].str.replace('pt', '')
a['team'] = a['team'].str.replace('.', '')
a['team'] = a['team'].str.strip()
a['date'] = match_date
a['date'] = pd.to_datetime(a['date'],format = '%y%b%d')
a = a.rename(columns = {'off.':'off','def.':'def'})

a.to_csv(fr'{full_path}\{current_season}_{current_season+1}_{match_date}.csv')
```

```python
### SPI SCRAPER FOR HISTORICAL RESULTS

current_season = int(input('What is the current season?'))


url = r'https://projects.fivethirtyeight.com/soccer-predictions/premier-league/'

driver = webdriver.Chrome()

r = driver.get(url) # proxies=proxies
time.sleep(2)

date_dict = {'Sept':'Sep','March':'Mar','April':'Apr','June':'Jun','July':'Jul'}

for j in range (2021,current_season + 1):
    
    if not os.getcwd().endswith('Football Forecasting Version 2'):
        os.chdir(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))

    current_folder = os.getcwd()

    relative_folder = fr'{j}_{j+1}_SPIdata'   

    full_path = os.path.join(current_folder, relative_folder)

    isExist = os.path.exists(full_path)


    if not isExist:
        os.makedirs(full_path)

    xpath_filter_by_season = r'//*[@id="season-selector"]/div[2]/select'
    driver.find_element("xpath", xpath_filter_by_season).click()
    time.sleep(2)

    xpath_season_i = fr'//*[@id="season-selector"]/div[2]/select/option[{-j + 2023}]' #
    driver.find_element("xpath", xpath_season_i).click()
    time.sleep(2)

    for i in range(1,39):

        xpath_filter_by_matchweek = r'//*[@id="forecast-selector"]/div[2]/select'
        driver.find_element("xpath", xpath_filter_by_matchweek).click()
        time.sleep(2)

        try:

            xpath_matchweek_i = fr'//*[@id="forecast-selector"]/div[2]/select/option[{i}]' #
            driver.find_element("xpath", xpath_matchweek_i).click()
            time.sleep(2)

            source=driver.page_source
            soup = BeautifulSoup(source,'html.parser') 
            table_soup = soup.find('option', {'value':i-1})
            match_date = table_soup.get_text().upper().replace('. ','')
            for key,value in date_dict.items():
                match_date = match_date.replace(key.upper(),value.upper())
            match_date = match_date.replace(' ','')


            for k in ['AUG','SEP','OCT','NOV','DEC']:
                if k in match_date:
                    match_date = str(j)[-2:] + match_date
                    break
                elif match_date == 'TODAY':
                    match_date = date.today().strftime('%y%b%d')
                    break
                elif k not in match_date and k == 'DEC':
                    match_date =str(j + 1)[-2:] + match_date
                else:
                    continue

            dfs = pd.read_html(driver.page_source)
            a = dfs[0].replace(' ',':')
            a.columns = a.columns.droplevel()
            a = a[['team','spi','off.','def.']]
            a['team'] = a['team'].str.replace('\d+', '',regex = True)
            a['team'] = a['team'].str.replace('pts', '')
            a['team'] = a['team'].str.replace('pt', '')
            a['team'] = a['team'].str.replace('.', '')
            a['team'] = a['team'].str.strip()
            a['date'] = match_date
            a['date'] = pd.to_datetime(a['date'],format = '%y%b%d')
            a = a.rename(columns = {'off.':'off','def.':'def'})

            a.to_csv(fr'{full_path}\{j}_{j+1}_{match_date}.csv')
        except:
            print(i)
            continue

        
```
